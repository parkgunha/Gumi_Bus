### ***노선번호목록(노선의 번호 = 버스번호) 조회***

```python
import requests

url = 'http://apis.data.go.kr/1613000/BusRouteInfoInqireService/getRouteNoList'
params ={'serviceKey' : 'wRGALjymSzbi/GLVDdu69u7/jHfZ6d1luMJ6H3t3uCt5BETeM0wzS48xVuJ0Gv8lqct7bqmHnr6jA47/T4OvLA=='
         , 'numOfRows': '1000000', '_type' : 'json', 'cityCode' : '37050'}

response = requests.get(url, params=params)
print(response.content)
```

```python
import json

data = json.loads(response.text)

print(data)
print(type(data))
print(data.keys())
```

```python
# 총 노선 번호 개수는 341개
data['response']['body']['totalCount']
```

```python
data['response']['body']
```

```python
# json 파일 -> 딕셔너리. 딕셔너리는 [key]를 통해 value를 얻을 수 있고, 리스트는 [0]처럼 슬라이싱이 가능
# 리스트에 여러 개의 딕셔너리가 있을 때는 딕셔너리의 키가 컬럼이 되고 value가 각 컬럼의 값이 됨

data['response']['body']['items']['item'][0]
```

```python
# Json파일을 DF로 만들기
import pandas as pd

df = pd.DataFrame(data['response']['body']['items']['item'])
df.head(3)
```

```python
# 데이터 가독성을 위해 컬럼명을 한글로 변환
df = df.rename(columns={'endnodenm': '종점', 'endvehicletime': '막차시간', 'routeid': '노선ID', 'routeno': '노선번호', 'routetp': '노선유형'
                   ,'startnodenm': '기점', 'startvehicletime': '첫차시간'})
df.head(3)
```

```python
# 910번 버스의 노선 정보 확인 -> 네이버 지도와 일치, 구미시 버스정보시스템과도 일치 -> 최신 버전임
df[df.노선번호 == 910]
```

```python
# null값 확인
df.info()
```

```python
# null값 확인 -> 13번 버스는 운행되고 있지 않기 때문에 첫차시간과 막차시간이 없음
display(df[df.막차시간.isna()], df[df.첫차시간.isna()])
```

```python
# null값을 삭제: 운행하지 않고 있는 노선이기 때문에
print(df.shape)
df = df.dropna()
print(df.shape)
```

```python
# df를 csv로 내보내기-> 다른 파일에서 노선 번호를 돌면서 api 받기
df.to_csv("노선번호.csv", index=False)
```
